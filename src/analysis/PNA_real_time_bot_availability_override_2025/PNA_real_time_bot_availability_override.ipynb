# Starburst Data

```python
QUERY_NAME = "test_starburst_exposures.sql"
START_DATE = "'2025-01-01'"
END_DATE = "'2025-04-01'"

params = {"start_date": str(START_DATE), "end_date": str(END_DATE)}

df = q.run_query_starburst(  # or run_table_explorer_bigquery
    QUERY_NAME, params=params, csv_file=QUERY_NAME, load_csv_file=True
)

df.head()
```

```python
pivot_params = {
    "df": df,
    "index": [1],
    "columns": [2],
    "values": [3],
    "aggfuncs": ['sum'],
}

dfp = dataframe_tools.Pivot(df, pivot_params).run_pivot()
dfp
```

```python
figure_params = {
    "n_plots": 1,
    "fig_length": 10,
    "fig_height": 6,
    "x_rotation": 45,
    "share_x": False,
    "share_y": False,
    "x_limits": None,  # [(X, X)]
    "y_limits": None,  # [(X, X)]
    "log_axis": None,  # ['x', 'y', 'both']
    "title": 'Daily customer PNA allocation - Starburst',
}

# Lineplot
plot_params = [
    {
        "plot_type": "lineplot",
        "x_column": [1],
        "y_columns": [2, 3],
        "colors": ["b", "o"],  # ['o','g','r','pu','br','pi','gr','ol','cy']
        "styles": ["-", "-"],  # ['--'. '-', ':']
        "legend": True,
    },
]

dataframe_tools.Plotter(
    dfp, plot_params=plot_params, figure_params=figure_params
).run_plotter()
```
